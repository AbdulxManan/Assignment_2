```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Data=pd.read_csv('K1.csv')
Data
```

```python
counter=Data['data'].count()
k=2;
Data['Cluster']=0
centre_value_1=13;
centre_value_2=55;
iteration_number=0
Data[centre_value_1]=0;
Data[centre_value_2]=0;
Data['Group-A',iteration_number]=0
Data['Group-B',iteration_number]=0
Data
y=['10', '15', '20', '25', '30', '35', '40', '45', '50', '55', '60', '65' ,'70' ,'75']
plt.scatter(Data['data'],y,c=Data.Cluster)
```

```python
def difference():
    for i in range(counter):
        temp1=((Data['data'].iloc[i])-centre_value_1)
        temp2=((Data['data'].iloc[i])-centre_value_2)
        temp1=(temp1)**2
        temp2=(temp2)**2
        temp1=(temp1)**0.5
        temp2=(temp2)**0.5
        Data[centre_value_1].iloc[i]=temp1;
        Data[centre_value_2].iloc[i]=temp2;
        temp1=Data[centre_value_1].iloc[i]
        temp2=Data[centre_value_2].iloc[i]
        if temp1<temp2:
            extra=Data['data'].iloc[i]
            Data['Group-A',iteration_number].iloc[i]=extra
        else:
            extra=Data['data'].iloc[i]
            Data['Group-B',iteration_number].iloc[i]=extra
            
```

```python
difference()
```

```python
while True:    
    extra1=Data['Group-A',iteration_number].sum()
    extra2=Data['Group-B',iteration_number].sum()
    div1=0
    div2=0
    for x in range(counter):
        if Data['Group-A',iteration_number].iloc[x]!=0:
            div1=div1+1
        if Data['Group-B',iteration_number].iloc[x]!=0:
            div2=div2+1
    avg1=extra1/div1
    avg2=extra2/div2
    avg1=int(avg1)
    avg2=int(avg2)
    if(avg1==centre_value_1):
        if(avg2==centre_value_2):
            break
    centre_value_1=avg1
    centre_value_2=avg2
    Data[centre_value_1]=0
    Data[centre_value_2]=0
    iteration_number=iteration_number+1
    Data['Group-A',iteration_number]=0
    Data['Group-B',iteration_number]=0
    difference()
```

```python
for x in range(counter):
    temp=Data['Group-A',iteration_number].iloc[x]
    if temp!=0:
        Data['Cluster'].iloc[x]=1
    else:
        Data['Cluster'].iloc[x]=2

Data
```
